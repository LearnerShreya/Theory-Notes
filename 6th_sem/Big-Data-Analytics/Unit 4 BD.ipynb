```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import squarify

# Data
values = [10, 20, 20, 30, 40]
categories = ['A', 'B', 'C', 'D', 'E']

# Setup figure
fig, axs = plt.subplots(3, 4, figsize=(18, 12))
axs = axs.flatten()

# 1. Bar Chart
axs[0].bar(categories, values)
axs[0].set_title("Bar Chart")

# 2. Pie Chart
axs[1].pie(values, labels=categories, autopct='%1.1f%%')
axs[1].set_title("Pie Chart")

# 3. Treemap
squarify.plot(sizes=values, label=categories, ax=axs[2])
axs[2].set_title("Treemap")
axs[2].axis('off')

# 4. Line Chart
axs[3].plot(values, marker='o')
axs[3].set_title("Line Chart")

# 5. Scatter Plot
axs[4].scatter(values, values[::-1])
axs[4].set_title("Scatter Plot")

# 6. Box Plot
axs[5].boxplot(values)
axs[5].set_title("Box Plot")

# 7. Histogram
axs[6].hist(values, bins=5)
axs[6].set_title("Histogram")

# 8. Heatmap
matrix_data = np.array([10, 20, 20, 30, 40, 50]).reshape(2, 3)
sns.heatmap(matrix_data, annot=True, fmt="d", ax=axs[7], cmap="YlGnBu")
axs[7].set_title("Heatmap")

# 9. Simulated Geo Map as placeholder
axs[8].bar(categories, values, color='green')
axs[8].set_title("Geo Map (Simulated)")

# Hide unused subplots
for i in range(9, len(axs)):
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```
