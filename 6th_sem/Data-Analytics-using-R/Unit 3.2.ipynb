## **1. Introduction to Data Visualization in R**

### **What is Data Visualization?**
Data visualization is the process of transforming raw data into graphical representations such as charts and graphs. This makes it easier to identify trends, patterns, and insights that are otherwise difficult to interpret from raw data.

### **Why is Data Visualization Important?**
- **Improves Data Interpretation:** Makes complex data easier to understand.
- **Enhances Decision Making:** Helps in making data-driven decisions quickly.
- **Identifies Trends & Patterns:** Provides insights into data distribution and relationships.
- **Detects Outliers:** Helps in identifying unusual data points.
- **Summarizes Large Datasets:** Condenses large amounts of information into a simple visual representation.

### **Types of Charts in R**
R provides various types of charts and graphs for data visualization. Some of the most commonly used ones include:
- **Histogram** - Used for understanding the frequency distribution of data.
- **Boxplot** - Used for detecting outliers and visualizing spread.
- **Bar Chart** - Used for comparing different categories.
- **Line Graph** - Used for displaying trends over time.
- **Scatter Plot** - Used for showing relationships between two variables.
- **Pie Chart** - Used for showing proportions of a whole.

| Chart Type  | Purpose | Data Type | Best For | Function Used |
|------------|---------|----------|----------|--------------|
| **Histogram** | Shows frequency distribution | Continuous numerical | Understanding data distribution | `hist()` |
| **Boxplot** | Displays summary statistics & outliers | Continuous numerical | Comparing distributions across groups | `boxplot()` |
| **Bar Chart** | Compares different categories | Categorical | Comparing different groups | `barplot()` |
| **Line Graph** | Displays trends over time | Time series data | Tracking changes over periods | `plot()` |
| **Scatter Plot** | Shows relationships between two variables | Continuous numerical | Identifying correlations & clusters | `plot()` |
| **Pie Chart** | Represents proportions of a whole | Categorical | Showing percentage distribution | `pie()` |

# **3. Types of Charts and Graphs in R (With Code & Examples)**

## **3.1 Histogram (`hist()`)**

### **What is a Histogram?**
A histogram is a graphical representation that shows how data is distributed. It divides the data into bins (intervals) and represents the frequency of values in each bin using bars.

### **Key Features of a Histogram:**
- It helps in understanding the distribution of numerical data.
- The x-axis represents data intervals (bins), and the y-axis represents frequency.
- The height of each bar shows how many data points fall within that range.

### **When to Use a Histogram?**
- To analyze the distribution of test scores.
- To check the frequency of customer purchases.
- To identify the spread of data values in a dataset.

### **R Code Example**
```r
hist(mtcars$mpg, 
     col="blue", 
     main="Histogram of Miles Per Gallon", 
     xlab="Miles Per Gallon", 
     breaks=10)
```
 **Reference:** [Histogram in R](https://www.r-graph-gallery.com/histogram.html)

---

## **3.2 Boxplot (`boxplot()`)**

### **What is a Boxplot?**
A boxplot is a standardized way of displaying data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It also helps in detecting outliers.

### **Key Features of a Boxplot:**
- Displays the median and quartiles of the dataset.
- Identifies the spread and symmetry of data.
- Highlights outliers using points outside the whiskers.

### **When to Use a Boxplot?**
- To compare salary distributions in different companies.
- To analyze student performance variation in exams.
- To check variations in product prices across different stores.

### **R Code Example**
```r
boxplot(mtcars$mpg, 
        col="red", 
        main="Boxplot of MPG", 
        ylab="Miles Per Gallon")
```
 **Reference:** [Boxplot in R](https://www.r-graph-gallery.com/boxplot.html)

---

## **3.3 Bar Chart (`barplot()`)**

### **What is a Bar Chart?**
A bar chart represents categorical data using rectangular bars. The length of each bar is proportional to the value it represents.

### **Key Features of a Bar Chart:**
- The bars can be displayed vertically or horizontally.
- It is used to compare different categories.
- Each bar represents a distinct group in the dataset.

### **When to Use a Bar Chart?**
- To compare product sales across different stores.
- To analyze population distribution by country.
- To examine responses in a survey.

### **R Code Example**
```r
counts <- table(mtcars$cyl)
barplot(counts, 
        col="green", 
        main="Cylinder Count in Cars", 
        xlab="Cylinders", 
        ylab="Frequency")
```
 **Reference:** [Bar Chart in R](https://www.r-graph-gallery.com/barplot.html)

---

## **3.4 Line Graph (`plot()`)**

### **What is a Line Graph?**
A line graph is used to display information as a series of data points connected by straight lines. It is often used to show trends over time.

### **Key Features of a Line Graph:**
- Best for time-series data analysis.
- Helps in identifying increasing or decreasing trends.
- Can compare multiple series in one graph.

### **When to Use a Line Graph?**
- To track temperature changes over time.
- To observe stock price trends.
- To monitor website traffic over months.

### **R Code Example**
```r
time <- 1:10
data <- c(5, 7, 9, 12, 15, 18, 22, 26, 30, 35)
plot(time, data, 
     type="o", 
     col="blue", 
     main="Line Graph Example", 
     xlab="Time", 
     ylab="Values")
```
 **Reference:** [Line Graph in R](https://www.r-graph-gallery.com/line-chart.html)

## **3.5 Scatter Plot (`plot()`)**  

### **What is a Scatter Plot?**  
A scatter plot is used to display relationships between two continuous variables. Each point on the plot represents a pair of values.  

### **Key Features of a Scatter Plot:**  
- Used for correlation analysis between two variables.  
- Points are scattered across the plot, showing patterns or clusters.  
- Helps detect linear and non-linear relationships.  

### **When to Use a Scatter Plot?**  
- To analyze the relationship between height and weight.  
- To observe the correlation between advertising expenses and sales.  
- To identify trends between temperature and electricity consumption.  

### **R Code Example**  
```r
plot(mtcars$wt, mtcars$mpg,  
     col="blue",  
     main="Scatter Plot of Weight vs MPG",  
     xlab="Car Weight",  
     ylab="Miles Per Gallon",  
     pch=16)  # pch sets point style
```  
 **Reference:** [Scatter Plot in R](https://www.r-graph-gallery.com/scatterplot.html)

## **3.6 Pie Chart (`pie()`)**  

### **What is a Pie Chart?**  
A pie chart represents proportions of different categories as slices of a circle. Each slices size corresponds to its percentage of the total.  

### **Key Features of a Pie Chart:**  
- Displays categorical data as percentages.  
- Useful for understanding composition or distribution.  
- Limited to small datasets for clear readability.  

### **When to Use a Pie Chart?**  
- To analyze market share of different brands.  
- To compare department-wise expenses in a company.  
- To show percentage distribution of survey responses.  

### **R Code Example**  
```r
slices <- c(30, 20, 25, 15, 10)  
labels <- c("A", "B", "C", "D", "E")  
pie(slices, labels=labels,  
    col=rainbow(length(slices)),  
    main="Pie Chart Example")  
```  
 **Reference:** [Pie Chart in R](https://www.r-graph-gallery.com/piechart.html)

## **3.7 Heatmap (`heatmap()`)**  

### **What is a Heatmap?**  
A heatmap represents data in matrix form using color gradients, where darker shades indicate higher values.  

### **Key Features of a Heatmap:**  
- Used for visualizing patterns in large datasets.  
- Helps in correlation analysis between multiple variables.  
- Colors indicate variations in magnitude.  

### **When to Use a Heatmap?**  
- To analyze customer behavior across different hours of the day.  
- To observe gene expression in biological studies.  
- To visualize correlation matrices in finance.  

### **R Code Example**  
```r
data_matrix <- as.matrix(mtcars)  
heatmap(data_matrix,  
        col=heat.colors(256),  
        main="Heatmap of Car Dataset")  
```  
 **Reference:** [Heatmap in R](https://www.r-graph-gallery.com/heatmap.html)

## **3.8 Density Plot (`density()`)**  

### **What is a Density Plot?**  
A density plot is a smoothed version of a histogram that estimates the probability distribution of a continuous variable.  

### **Key Features of a Density Plot:**  
- Provides a smooth curve instead of discrete bins.  
- Useful for comparing multiple distributions.  
- Helps in understanding the shape of data distribution.  

### **When to Use a Density Plot?**  
- To visualize the distribution of test scores.  
- To compare income distributions across different groups.  
- To analyze stock price variations.  

### **R Code Example**  
```r
density_data <- density(mtcars$mpg)  
plot(density_data,  
     col="purple",  
     main="Density Plot of MPG",  
     xlab="Miles Per Gallon")  
```  
 **Reference:** [Density Plot in R](https://www.r-graph-gallery.com/density-plot.html)

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data
data = [5, 10, 10, 15, 20, 25, 25, 30]

# Set up the figure and axes
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# 1. Histogram
axs[0, 0].hist(data, color='lightblue', edgecolor='black')
axs[0, 0].set_title('Histogram')
axs[0, 0].set_xlabel('Values')
axs[0, 0].set_ylabel('Frequency')

# 2. Boxplot
sns.boxplot(data=data, ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title('Boxplot')
axs[0, 1].set_ylabel('Values')

# 3. Bar Chart
axs[0, 2].bar(np.unique(data), [data.count(i) for i in np.unique(data)], color='lightcoral')
axs[0, 2].set_title('Bar Chart')
axs[0, 2].set_xlabel('Values')
axs[0, 2].set_ylabel('Frequency')

# 4. Line Graph
axs[1, 0].plot(data, marker='o', color='blue')
axs[1, 0].set_title('Line Graph')
axs[1, 0].set_xlabel('Index')
axs[1, 0].set_ylabel('Values')

# 5. Scatterplot
axs[1, 1].scatter(data, data, color='purple')
axs[1, 1].set_title('Scatterplot')
axs[1, 1].set_xlabel('X Values')
axs[1, 1].set_ylabel('Y Values')

# 6. Pie Chart
axs[1, 2].pie([data.count(i) for i in np.unique(data)], labels=np.unique(data), colors=plt.cm.rainbow(np.linspace(0, 1, len(np.unique(data)))))
axs[1, 2].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Save the figure as an image
plt.savefig('visualization.png')

# Show the plot
plt.show()
````
