# **8. Bayes' Rule and Posterior Probability**

### **Bayes' Rule**:
Bayes' Theorem allows you to update the probability of a hypothesis based on new evidence.

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

Where:
- \( P(A|B) \) is the **posterior probability** (probability of hypothesis \( A \) given the evidence \( B \)).
- \( P(B|A) \) is the **likelihood** (probability of evidence \( B \) given \( A \)).
- \( P(A) \) is the **prior probability** (probability of \( A \) before seeing the evidence).
- \( P(B) \) is the **marginal likelihood** (probability of evidence \( B \)).

### **Posterior Probability**:
Posterior is the updated probability after considering the new data.

### **Example Code:**
```R
# Example: Updating the probability of a disease given a positive test result
P_disease <- 0.01  # Prior probability (1% chance of having the disease)
P_positive_given_disease <- 0.9  # Likelihood (90% chance of positive test if diseased)
P_positive <- 0.05  # Probability of a positive test result

P_disease_given_positive <- (P_positive_given_disease * P_disease) / P_positive
print(P_disease_given_positive)
```

# **9. Binomial Distribution, Mass Function, and Cumulative Distribution Function**

### **Binomial Distribution**:
The binomial distribution describes the number of successes in a fixed number of independent trials, each with two possible outcomes (success or failure).

- **Probability Mass Function (PMF)**:
  \[
  P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
  \]
  Where:
  - \( n \) is the number of trials.
  - \( k \) is the number of successes.
  - \( p \) is the probability of success.

- **Cumulative Distribution Function (CDF)**:
  \[
  P(X \leq k) = \sum_{i=0}^{k} P(X = i)
  \]

### **R Functions**:
- `dbinom(x, size, prob)` â€“ PMF
- `pbinom(q, size, prob)` â€“ CDF
- `rbinom(n, size, prob)` â€“ Generate random binomial variables

### **Example Code:**
```R
# Binomial Distribution example: 5 trials, probability of success = 0.6
n <- 5  # Number of trials
p <- 0.6  # Probability of success

# Probability of 3 successes
prob_3_successes <- dbinom(3, n, p)
print(prob_3_successes)

# Cumulative probability of 3 or fewer successes
cum_prob_3 <- pbinom(3, n, p)
print(cum_prob_3)
```

# **10. CDF and PDF of a Normal Distribution**

### **Normal Distribution**:
A normal distribution is a continuous probability distribution characterized by the bell curve, with the mean (\( \mu \)) and standard deviation (\( \sigma \)).

- **Probability Density Function (PDF)**:
  \[
  f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
  \]

- **Cumulative Distribution Function (CDF)**:
  \[
  F(x) = \int_{-\infty}^{x} f(t) dt
  \]

### **R Functions**:
- `dnorm(x, mean, sd)` â€“ PDF
- `pnorm(q, mean, sd)` â€“ CDF
- `qnorm(p, mean, sd)` â€“ Quantile function
- `rnorm(n, mean, sd)` â€“ Random sample from normal distribution

### **Example Code:**
```R
# Normal Distribution: mean = 0, sd = 1
mean <- 0
sd <- 1

# PDF at x = 1
pdf_value <- dnorm(1, mean, sd)
print(pdf_value)

# CDF at x = 1
cdf_value <- pnorm(1, mean, sd)
print(cdf_value)
```

# **12. Z-Test**

### **Definition:**
A Z-test is used to determine whether there is a significant difference between the sample mean and population mean, given the population standard deviation.

### **Z-Test Formula**:
\[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
\]
Where:
- \( \bar{X} \) is the sample mean.
- \( \mu \) is the population mean.
- \( \sigma \) is the population standard deviation.
- \( n \) is the sample size.

### **R Function**:
- `z.test()` (from the `BSDA` package) or use the formula manually.

### **Example Code:**
```R
# Install package
install.packages("BSDA")
library(BSDA)

# Sample data
sample_data <- c(50, 52, 53, 54, 50, 48, 51)
z.test(sample_data, mu = 50, sigma.x = 2, conf.level = 0.95)
```

# **14. Confidence Intervals**

### **Definition:**
A confidence interval provides an estimated range of values for an unknown population parameter, with a given level of confidence.

- **Formula for Confidence Interval (CI)**:
  \[
  CI = \bar{X} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
  \]

Where:
- \( \bar{X} \) is the sample mean.
- \( Z_{\alpha/2} \) is the critical value for the given confidence level.
- \( \sigma \) is the sample standard deviation.
- \( n \) is the sample size.

### **R Function**:
- `confint(model)` â€“ Confidence intervals for regression models.

### **Example Code:**
```R
# Confidence Interval for Mean of mpg
t.test(mtcars$mpg, conf.level = 0.95)
```

# **15. F Test**

### **Definition:**
An F-test is used to compare two variances to see if they are significantly different.

- **Formula for F-statistic**:
  \[
  F = \frac{\sigma_1^2}{\sigma_2^2}
  \]

Where:
- \( \sigma_1^2 \) and \( \sigma_2^2 \) are the sample variances.

### **R Function**:
- `var.test(x, y)` â€“ Performs an F-test for equality of variances.

### **Example Code:**
```R
# F-test to compare variances of two samples
f_test <- var.test(mtcars$mpg, mtcars$hp)
print(f_test)
```

# **16. Chi-Square Test**

### **Definition:**
A Chi-Square test is used to determine if there is a significant association between categorical variables.

- **Chi-Square Statistic**:
  \[
  \chi^2 = \sum \frac{(O - E)^2}{E}
  \]
  Where:
  - \( O \) is the observed frequency.
  - \( E \) is the expected frequency.

### **R Function**:
- `chisq.test(x, y)` â€“ Performs a Chi-Square test.

### **Example Code:**
```R
# Chi-Square Test: Association between two categorical variables
observed <- matrix(c(10, 20, 20, 40), nrow = 2)
chisq_test <- chisq.test(observed)
print(chisq_test)
```
