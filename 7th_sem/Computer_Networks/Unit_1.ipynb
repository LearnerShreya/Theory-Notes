## üåê **1. Representation of Data and Its Flow**

* **Data Representation**: Information must be represented in a form suitable for transmission.

  * **Text** ‚Üí ASCII (7 bits), Unicode (16/32 bits).
  * **Numbers** ‚Üí Binary, Decimal, Hexadecimal.
  * **Images** ‚Üí Pixels, JPEG, GIF, PNG.
  * **Audio** ‚Üí MP3, WAV (analog ‚Üí digital using sampling).
  * **Video** ‚Üí MPEG, MP4 (frames per second).

* **Data Flow Types**:

  1. **Simplex** ‚Üí One-way only (e.g., keyboard ‚Üí monitor).
  2. **Half-Duplex** ‚Üí Both ways, but one at a time (e.g., walkie-talkie).
  3. **Full-Duplex** ‚Üí Both ways simultaneously (e.g., telephone).

---

## üåê **2. Computer Networks**

* **Definition**: A **network** is a collection of interconnected devices (computers, servers, etc.) that share resources and exchange data.
* **Purposes**: Sharing files, printers, internet, communication, reliability, cost saving.

### **Types of Networks**

1. **LAN (Local Area Network)** ‚Üí small area (college, office).
2. **MAN (Metropolitan Area Network)** ‚Üí city-wide (ISP networks).
3. **WAN (Wide Area Network)** ‚Üí covers countries/continents (Internet).

---

## üåê **3. Various Connection Topologies**

A **topology** = physical/logical structure of how devices are connected.

1. **Bus Topology**

   * Single backbone cable, devices attached.
   * ‚úÖ Easy, cheap. ‚ùå Collisions, difficult fault detection.

2. **Star Topology**

   * Devices connected to a central hub/switch.
   * ‚úÖ Easy to manage, reliable. ‚ùå Hub failure = whole network down.

3. **Ring Topology**

   * Devices connected in a circle, data travels one direction.
   * ‚úÖ Predictable. ‚ùå Failure of one node affects whole network.

4. **Mesh Topology**

   * Every device connected to every other.
   * ‚úÖ Reliable, no single point failure. ‚ùå Very expensive, complex.

5. **Hybrid Topology**

   * Combination of two or more (e.g., Star + Bus).

---

## üåê **4. Protocols and Standards**

* **Protocol** = Set of **rules** for communication (like grammar for computers).

  * Defines **syntax (format), semantics (meaning), timing**.
* **Standards** ensure **interoperability** among different manufacturers.

### **Major Organizations**

* **ISO (International Organization for Standardization)** ‚Üí OSI Model.
* **IEEE (Institute of Electrical and Electronics Engineers)** ‚Üí LAN/Wi-Fi.
* **IETF (Internet Engineering Task Force)** ‚Üí Internet standards.
* **ITU (International Telecommunication Union)** ‚Üí Telecom standards.

---

## üåê **5. OSI Model (7 Layers)**

OSI = **Open Systems Interconnection** model. Provides a layered framework for communication.

| Layer           | Function                               | Example             |
| --------------- | -------------------------------------- | ------------------- |
| 7. Application  | User interaction                       | HTTP, FTP, DNS      |
| 6. Presentation | Translation, compression, encryption   | JPEG, SSL           |
| 5. Session      | Dialog control, synchronization        | NetBIOS             |
| 4. Transport    | End-to-end delivery, error control     | TCP, UDP            |
| 3. Network      | Logical addressing, routing            | IP, ICMP            |
| 2. Data Link    | Node-to-node delivery, error detection | Ethernet, PPP       |
| 1. Physical     | Transmission of raw bits               | Cables, radio waves |

üëâ Easy trick to remember: **All People Seem To Need Data Processing** (from top to bottom).

## üåê **6. Transmission Media**

Medium = physical path for transmitting data.

### (a) **Wired Media**

1. **Twisted Pair Cable** (STP, UTP) ‚Äì telephone, LAN.
2. **Coaxial Cable** ‚Äì cable TV, older LANs.
3. **Optical Fiber** ‚Äì very high speed, long distance, immune to EMI.

### (b) **Wireless Media**

* **Radio waves** ‚Äì Wi-Fi, Bluetooth.
* **Microwaves** ‚Äì satellite, cellular.
* **Infrared** ‚Äì remote controls, short range.

---

## üåê **7. LAN (Local Area Network)**

* **Definition**: Network within small area (home, office, campus).

### (a) **Wired LAN**

* Uses Ethernet cables, hubs, switches.
* High speed, reliable.

### (b) **Wireless LAN (WLAN)**

* Uses Wi-Fi (IEEE 802.11).
* Easy mobility, but less secure.

### (c) **Connecting LANs**

* **Switches** (connect LAN segments).
* **Routers** (connect LAN to WAN/Internet).

### (d) **Virtual LAN (VLAN)**

* Logical division of LANs (via switches).
* Used for **security, traffic management, segmentation**.

---

## üåê **8. Techniques for Bandwidth Utilization**

To **maximize channel efficiency** ‚Üí use **multiplexing**.

### **Multiplexing**

* Process of **combining multiple signals** into one channel.

#### 1. **FDM (Frequency Division Multiplexing)**

* Each signal ‚Üí different frequency band.
* Example: Radio, TV channels.

#### 2. **TDM (Time Division Multiplexing)**

* Each signal ‚Üí assigned a **time slot**.
* Example: Telephony, digital data.

#### 3. **WDM (Wavelength Division Multiplexing)**

* Optical fibers ‚Üí multiple light wavelengths (colors).
* Example: Fiber-optic backbone.

---

## üåê **9. Spread Spectrum Concepts**

Spread spectrum = **spreading signals over wide frequency range** for security & noise resistance.

1. **FHSS (Frequency Hopping Spread Spectrum)**

   * Signal hops between frequencies in sequence.
   * Example: Bluetooth.

2. **DSSS (Direct Sequence Spread Spectrum)**

   * Data bits multiplied by longer pseudo-noise code.
   * Example: Wi-Fi (802.11b).

üëâ Provides **security, resistance to interference, and better reliability**.
