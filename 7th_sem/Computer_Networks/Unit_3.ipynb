## âš¡ **2. Functions of Network Layer**

1. **Logical Addressing**

   * Adds an **IP address** (IPv4 or IPv6) to identify each device uniquely.
   * Ensures **host-to-host delivery**.

2. **Routing**

   * Finds the **best path** for packet delivery using routing algorithms (e.g., OSPF, RIP, BGP).

3. **Packet Forwarding**

   * Routers read the **destination address** in the packet header and forward it to the next hop.

4. **Switching**

   * Uses **switching techniques** (circuit, packet, message) for data transfer.

5. **Fragmentation & Reassembly**

   * Breaks large packets into smaller fragments if the data exceeds the **MTU (Maximum Transmission Unit)**.
   * Reassembles them at the destination.

6. **Error Handling & Congestion Control**

   * Uses protocols like **ICMP** for error reporting.
   * Manages **traffic congestion**.

Switching = the **method of transferring data** from sender to receiver through the network.
There are **three main types**:

### 3.1 **Circuit Switching**

* A **dedicated communication path** is established between sender and receiver before data transfer.
* Resources remain reserved until communication ends.

### 3.2 **Packet Switching**

* Data is divided into **small packets**.
* Each packet is sent independently through the network.
* At the destination, packets are **reassembled** in order.

### 3.3 **Message Switching**

* Entire message is treated as a **single block** and transferred from node to node.
