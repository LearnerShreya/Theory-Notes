A network is a setup where multiple devices are linked together so they can share information, resources, and services. It includes end devices like computers and phones, networking devices like switches and routers, and the wired or wireless connections that let data move between them. The design of how these parts connect and communicate is called network architecture.

### Network diagram

### Network Devices


**Switch:**  
A switch connects multiple devices within the same network and forwards data only to the exact device it is intended for. This makes communication faster and reduces unnecessary traffic.

**Router:**  
A router connects different networks, directs data between them, and chooses the best path for data to reach its destination. It is what connects your local network to the internet.

## **Part 1: Basics of Networking**

### 1. **Basics of Network Architecture**

* **Definition:** Network architecture is the design and structure of a computer network that defines how devices (computers, routers, switches, servers) communicate.
* **Types:**

  * **ClientServer:** Clients request services, and servers provide them (e.g., web browsing).
  * **Peer-to-Peer (P2P):** Each device can act as both client and server (e.g., file sharing).
  * **Centralized:** One central device manages traffic.
  * **Distributed:** No single failure point, workload shared across multiple systems.
* **Key Elements:**

  * **End devices:** Computers, phones, printers.
  * **Networking devices:** Switches, routers, firewalls.
  * **Transmission media:** Wired (Ethernet cables) or wireless (Wi-Fi).

---

### 2. **Enterprise Network Constructs**

* Used in **companies/organizations** where multiple departments, branches, and users need secure and fast connectivity.
* **Components:**

  * **LAN (Local Area Network):** Small network (inside office/building).
  * **WAN (Wide Area Network):** Connects different LANs across cities/countries.
  * **WLAN (Wireless LAN):** Wi-Fi-based LAN.
  * **VPN (Virtual Private Network):** Secure connection over the internet.
  * **Data Center Networks:** Centralized servers and storage.
* **Goals:** Scalability, security, high performance, redundancy.

---

### 3. **Ethernet Framing**

* Ethernet is the most common **LAN technology**.
* **Ethernet Frame:** A structure that holds data before sending it across the network.
* **Main parts:**

  * **Preamble:** Synchronization bits.
  * **Destination MAC Address:** Receivers physical address.
  * **Source MAC Address:** Senders physical address.
  * **Type/Length:** Tells which protocol (IPv4, IPv6, ARP, etc.).
  * **Payload (Data):** Actual data being transmitted.
  * **Frame Check Sequence (FCS):** Error detection.
* **Purpose:** Ensures data is properly packaged for delivery.

---

### 4. **IP Addressing**

* **Definition:** A unique number assigned to each device in a network.
* **Types:**

  * **IPv4:** 32-bit, written as `192.168.1.1` (about 4.3 billion addresses).
  * **IPv6:** 128-bit, written as `2001:0db8:85a3::8a2e:0370:7334` (virtually unlimited).
* **Classes of IPv4:**

  * Class A: `1.0.0.0  126.255.255.255` (large networks).
  * Class B: `128.0.0.0  191.255.255.255` (medium networks).
  * Class C: `192.0.0.0  223.255.255.255` (small networks).
* **Public vs Private IP:**

  * **Public:** Accessible from the internet.
  * **Private:** Used inside organizations (e.g., `192.168.x.x`).
* **Subnetting:** Dividing a big network into smaller ones for better efficiency.

## **Part 2: Core Protocols and Data Movement**

---

### 5. **Internet Control Message Protocol (ICMP)**

* **Definition:** A network layer protocol used by devices (routers, hosts) to send **error messages and operational information**.
* **Purpose:** Helps in **diagnosing and troubleshooting** network issues.
* **Examples of use:**

  * **Ping:** Checks connectivity between two devices.
  * **Traceroute:** Shows the path taken by packets across the network.
* **Functions:**

  * Reports errors (e.g., destination unreachable).
  * Helps detect packet loss or delay.
* **Note:** ICMP is **not used for sending data**, only for control and error reporting.

---

### 6. **Address Resolution Protocol (ARP)**

* **Definition:** ARP maps an **IP address (logical)** to a **MAC address (physical)** in a LAN.
* **Why needed?** Because:

  * Computers communicate using **IP addresses**.
  * But data is delivered over Ethernet using **MAC addresses**.
* **Process:**

  1. A device wants to send data to IP `192.168.1.2`.
  2. It broadcasts an ARP request: Who has 192.168.1.2?
  3. The device with that IP replies with its MAC address.
  4. The sender saves the mapping in the **ARP table** for future use.
* **Issue:** ARP can be exploited in **ARP spoofing attacks** (security risk).

---

### 7. **Transport Layer Protocols**

* **Located at Layer 4 of OSI model.**
* **Main job:** Provide communication services between applications.

#### a) **TCP (Transmission Control Protocol)**

* **Connection-oriented:** Establishes a reliable connection before sending data.
* **Reliable:** Ensures data is delivered in order without errors.
* **Features:**

  * 3-way handshake (connection setup).
  * Flow control, congestion control.
  * Error checking.
* **Used for:** Web browsing (HTTP/HTTPS), Email (SMTP), File transfer (FTP).

#### b) **UDP (User Datagram Protocol)**

* **Connectionless:** No prior handshake.
* **Unreliable:** No guarantee of delivery/order.
* **Faster & lightweight** compared to TCP.
* **Used for:** Video streaming, Online gaming, VoIP (calls).

 **Key difference:**

* TCP = Reliable but slower.
* UDP = Fast but not reliable.

---

### 8. **Data Forwarding Scenario**

* **Definition:** The process by which a network device (like a router) sends a packet from **source to destination**.
* **Steps:**

  1. **Frame received** at routers interface.
  2. Router checks **destination IP** in the packet.
  3. Router looks up the **routing table** to find the best path.
  4. If next hop is found  packet is forwarded.
  5. If not found  packet is dropped (and ICMP error may be sent).
* **Forwarding uses:**

  * **MAC address** within a LAN.
  * **IP address** across different networks.

### 9. **Expanding the Enterprise Network**

* **Definition:** Process of scaling a companys network as it grows.
* **Ways to expand:**

  * Adding more **switches and routers**.
  * Using **redundancy** (backup links/devices).
  * Implementing **VLANs** to segment traffic.
  * Using **Load Balancers** for distributing traffic.
  * Adding **firewalls** and **VPNs** for security.
* **Goal:** Improve **scalability, performance, and reliability**.

### 10. **Navigating the CLI (Command-Line Interface)**

* **CLI:** A text-based interface used to configure network devices (switches, routers).
* **Why important?** GUI is limited; CLI gives **full control**.
* **Basic Modes in CLI:**

  * **User EXEC mode:** Limited commands (view only).
  * **Privileged EXEC mode:** Advanced commands (diagnostics).
  * **Global configuration mode:** Device-wide settings.
  * **Interface configuration mode:** Configure specific ports.
* **Examples:**

  * `ping 192.168.1.1`  check connectivity
  * `show ip route`  see routing table
  * `configure terminal`  enter config mode

---

### 11. **File System Navigation and Management**

* Network devices have their own **file systems** (like small storage).
* **Uses:**

  * Store **config files** (startup-config, running-config).
  * Store **OS images** (firmware).
  * Save **logs and backups**.
* **Common Commands:**

  * `dir`  list files.
  * `copy running-config startup-config`  save current config.
  * `delete`  remove files.

### 12. **VRP Operating System Image Management**

* **VRP (Versatile Routing Platform):** The OS used by **Huawei network devices**.
* Similar to **Cisco IOS**.
* **Why manage images?**

  * To **upgrade** device features.
  * To **fix bugs** or improve security.
* **Process:**

  1. Upload new OS image to device (via FTP/TFTP/USB).
  2. Verify image file (`display version`).
  3. Set the device to boot from new image.
  4. Reboot to activate.
* **Key Commands:**

  * `dir`  check files.
  * `boot-loader file-name`  set startup image.
  * `reboot`  restart with new image.

 *Flowchart  OS Image Upgrade:*

```
[Download Image]  [Verify File]  [Set as Boot Image]  [Reboot Device]
```
