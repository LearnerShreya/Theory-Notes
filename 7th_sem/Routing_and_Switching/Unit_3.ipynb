# **UNIT 3: NETWORK CONSTRUCTION**

**Network Connection:**
It is the link that allows computers and devices to communicate and share data or resources over a network, either through cables (wired) or radio signals (wireless).

---

## **1. Link Aggregation**

### **Definition**

Link Aggregation is the technique of combining multiple physical network links into one logical link to **increase bandwidth** and provide **redundancy**.

### **Key Points**

* Also known as **Port Trunking** or **EtherChannel**.
* Uses **IEEE 802.3ad (LACP – Link Aggregation Control Protocol)**.
* If one link fails, traffic automatically switches to remaining links.
* Works between **switches, routers, or servers**.

### **Benefits**

* **Higher throughput:** Combines multiple links’ speeds.
* **Load balancing:** Distributes traffic evenly.
* **Redundancy:** Prevents network downtime if one link fails.
* **Scalability:** Easy to expand by adding more links.

### **Example**

If each link has 1 Gbps capacity and 4 links are aggregated, total bandwidth = **4 Gbps**.

## **2. VLAN (Virtual Local Area Network) Principles**

### **Definition**

A **VLAN** is a logical network created within a physical network to **segment devices** for security, efficiency, or management reasons.

### **Key Concepts**

* VLANs are defined by **IEEE 802.1Q** standard.
* Each VLAN acts as a **separate broadcast domain**.
* VLANs are identified by a **VLAN ID (1–4094)**.

### **Types of VLAN**

1. **Default VLAN** – VLAN 1 on most switches; all ports belong to it initially.
2. **Data VLAN** – Carries user-generated traffic.
3. **Voice VLAN** – Dedicated to VoIP traffic.
4. **Management VLAN** – Used for switch management.
5. **Native VLAN** – Handles untagged traffic on trunk ports.

### **Advantages**

* Better **security** and **traffic control**.
* Reduces **broadcast traffic**.
* Simplifies **network management**.
* Allows **logical grouping** of users even if they are not physically close.

---

## **3. GARP (Generic Attribute Registration Protocol)** and **GVRP (GARP VLAN Registration Protocol)**

### **GARP**

* A protocol for **registering and deregistering attributes (like VLANs)** on network switches.
* Ensures consistent VLAN information across all devices.

### **GVRP**

* Built on top of GARP, defined by **IEEE 802.1Q**.
* Automatically **registers VLAN membership** on trunk links.
* Simplifies VLAN management in large networks by **automating propagation**.

### **How It Works**

* When a switch port joins a VLAN, it sends a **Join message**.
* Other switches learn this and add that VLAN to their ports automatically.

---

## **4. VLAN Routing**

### **Need**

Devices in different VLANs cannot communicate directly because VLANs are separate broadcast domains.
**Routing** is required for inter-VLAN communication.

### **Methods**

#### a. **Router-on-a-Stick**

* Uses one physical router interface with **sub-interfaces**, each assigned to a VLAN.
* Cost-effective for small networks.

#### b. **Layer 3 Switch**

* Performs routing directly at the switch level using **SVIs (Switch Virtual Interfaces)**.
* Faster than routers as switching is done in hardware.

### **Example**

* VLAN 10: 192.168.1.0/24
* VLAN 20: 192.168.2.0/24
* Router or L3 switch routes traffic between them.

## **5. Wireless LAN Overview**

### **Definition**

A **Wireless LAN (WLAN)** connects devices over radio waves instead of cables, allowing mobility within a local area.

### **Standards**

* Defined by **IEEE 802.11** family:

  * 802.11a/b/g/n/ac/ax (Wi-Fi 6).

### **Components**

1. **Access Point (AP):** Central device connecting wireless users to the wired network.
2. **Wireless Clients:** Devices with Wi-Fi adapters.
3. **Wireless Controller:** Manages multiple APs in enterprise setups.

### **Advantages**

* Mobility and flexibility.
* Easy installation and scalability.
* Ideal for places where cabling is hard or expensive.

### **Security Protocols**

* **WEP, WPA, WPA2, WPA3** – protect data and control access.

## **6. Bridging Enterprise Networks with Serial WAN Technology**

### **Definition**

Bridging with **Serial WAN (Wide Area Network)** technology allows enterprises to connect **remote LANs** over long distances using **serial communication links**.

### **Serial WAN Technologies**

* **HDLC (High-Level Data Link Control)**
* **PPP (Point-to-Point Protocol)**
* **Frame Relay**
* **ATM (Asynchronous Transfer Mode)**

### **Purpose**

* Extend enterprise LAN to branch offices.
* Enable communication between distant sites.

### **Process**

* Bridges connect two LANs.
* Serial interfaces on routers transmit data packets via **leased lines or WAN links**.

---

## **7. Frame Relay Principles**

### **Definition**

Frame Relay is a **packet-switched WAN technology** that sends data frames across virtual circuits.

### **Key Concepts**

* Works at **Data Link Layer (Layer 2)**.
* Uses **DLCIs (Data Link Connection Identifiers)** for logical addressing.
* Uses **Permanent Virtual Circuits (PVCs)** to establish fixed paths.

### **Advantages**

* Cost-effective for connecting multiple sites.
* Efficient bandwidth utilization.
* Less overhead than older WAN technologies.

### **Limitations**

* Replaced by modern broadband and MPLS networks.

---

## **8. Establishing DSL/ADSL Networks with PPPoE**

### **DSL (Digital Subscriber Line)**

Provides high-speed Internet over telephone lines.

### **ADSL (Asymmetric DSL)**

* Different upload and download speeds (download > upload).
* Common for home broadband.

### **PPPoE (Point-to-Point Protocol over Ethernet)**

* Combines **PPP features (authentication, encryption)** with **Ethernet**.
* Widely used for broadband Internet connections.

### **Process**

1. User’s modem connects to ISP via PPPoE session.
2. PPP handles **authentication (PAP/CHAP)**.
3. IP address assigned dynamically.

### **Advantages**

* Secure user authentication.
* Compatible with existing Ethernet and DSL infrastructure.

---

## **9. Network Address Translation (NAT)**

### **Definition**

NAT translates **private IP addresses** within a local network to a **public IP address** for Internet access.

### **Types of NAT**

1. **Static NAT:** One-to-one mapping between private and public IPs.
2. **Dynamic NAT:** Pool of public IPs shared among devices.
3. **PAT (Port Address Translation or NAT Overload):** Many devices share a single public IP using different ports.

### **Advantages**

* Conserves public IP addresses.
* Hides internal network structure (security).
* Enables Internet access for private networks.

---

## **10. Establishing Enterprise Radio Access Network (RAN) Solutions**

### **Definition**

A **Radio Access Network (RAN)** connects user devices (mobiles, IoT, laptops) to the core network using **radio signals**.

### **Components**

1. **Base Station (eNodeB / gNodeB):** Communicates with user devices.
2. **Controller (RNC / BSC):** Manages radio resources and handovers.
3. **Core Network:** Handles data routing, authentication, and switching.

### **Types**

* **2G/3G RAN:** GSM, UMTS.
* **4G RAN:** LTE.
* **5G RAN:** Uses **small cells** and **massive MIMO** for high speed.

### **Enterprise Use**

* Private 4G/5G networks for large campuses, factories, or smart cities.
* Enhanced security, low latency, and high throughput.
