### **1. Charts**

**Definition:**
A chart is a visual representation of data that helps to understand trends, comparisons, and relationships at a glance. It converts numbers into visual shapes that are easier to analyze.

**Why use charts?**
Because visuals help in faster understanding than raw data tables.

---

### **Types of Charts and Their Uses**

| **Chart Type**      | **Description**                                                                                                                         | **Best For**                                                                   | **Example**                                                                                   |
| ------------------- | --------------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------ | --------------------------------------------------------------------------------------------- |
| **Bar Chart**       | Uses rectangular bars (vertical or horizontal) to represent data.
