# **Unit 4 â€“ Part 1: The CCR Model (Definition of Target Objectives & Peer Groups)**

---

### **1. Introduction**

The **CCR Model** is one of the most important models in **Efficiency Measurement** â€” especially in **Data Envelopment Analysis (DEA)**.
It was introduced by **Charnes, Cooper, and Rhodes (1978)**, hence the name **CCR**.

The model helps to **measure the relative efficiency** of similar units (like departments, branches, companies, or machines) that use **multiple inputs** to produce **multiple outputs**.

---

### **2. What is Efficiency?**

In simple words,
$$\text{Efficiency} = \frac{\text{Output}}{\text{Input}}$$

But in real-life situations, we often have **many inputs and many outputs**, so we canâ€™t calculate it easily using just this formula.
Thatâ€™s where **DEA (Data Envelopment Analysis)** and the **CCR model** help.

---

### **3. What is the CCR Model?**

The **CCR model** is a **linear programming-based approach** that evaluates the **relative efficiency** of similar entities, called **Decision-Making Units (DMUs)**.

Each **DMU**:

* Uses **multiple inputs** (like labor, time, cost, materials)
* Produces **multiple outputs** (like sales, profit, services, performance)

The model finds the **best-performing DMUs** and then **compares others** with them.

---

### **4. Objective of the CCR Model**

The main goals are:

1. **Measure efficiency** of each DMU (organization, branch, machine, etc.)
2. **Identify benchmarks** â€” i.e., the most efficient DMUs (called *peer groups*).
3. **Set targets** for inefficient units to improve.
4. **Find reasons for inefficiency** â€” too much input, too little output, or both.
5. **Provide actionable recommendations** for performance improvement.

---

### **5. Key Concept â€“ Decision-Making Units (DMUs)**

**DMU** is any unit whose efficiency you want to evaluate.

Examples:

* Bank branches
* Hospitals
* Schools
* Retail stores
* Employees
* Machines

All DMUs must:

* Perform **similar functions**
* Use **similar types of inputs**
* Produce **similar types of outputs**

---

### **6. The Mathematical Idea (Conceptually)**

Each DMUâ€™s **efficiency score** is calculated as:

$$
\text{Efficiency} = \frac{\text{Weighted Sum of Outputs}}{\text{Weighted Sum of Inputs}}
$$

The model finds the **best possible weights** for inputs and outputs that give the DMU the highest efficiency (but without exceeding 1 for others).

So,

* Efficient DMUs have score = **1 (100%)**
* Inefficient DMUs have score **< 1**

---

### **7. Definition of Target Objectives**

The **target objectives** in the CCR model are basically the **goals or benchmarks** an inefficient DMU should achieve to become efficient.

If a DMU is not efficient (efficiency < 1), we determine:

* How much **inputs need to be reduced**, and
* How much **outputs need to be increased**

so that it reaches efficiency = 1.

**For example:**

* A bank branch uses 10 employees and produces 1000 transactions.
* An efficient branch uses the same 10 employees to produce 1200 transactions.
  So the target is to **increase output by 20%** to match efficiency.

This process defines **target objectives** for improvement.

---

### **8. Peer Groups**

A **Peer Group** (or **Reference Set**) consists of **efficient DMUs** that serve as **role models** for inefficient DMUs.

* Each inefficient DMU is compared against one or more **efficient DMUs**.
* These efficient ones are called **peers** or **benchmarks**.
* The model identifies **how much influence** each peer has in setting the target for the inefficient DMU.

**Example:**

* Suppose DMU A (inefficient) has efficiency 0.8.
* DMUs B and C are efficient.
* The model might show that DMU Aâ€™s performance should move toward a **combination of B (70%) and C (30%)** to become efficient.

That means **B and C form the peer group** for A.

---

### **9. Graphical Understanding (Simple View)**

Imagine a graph with:

* Inputs on the X-axis
* Outputs on the Y-axis

All efficient DMUs lie on the **Efficiency Frontier Line (Envelope Line)**.
All inefficient DMUs lie **below** this line.

The CCR model helps inefficient units **move up to the frontier** by setting **targets** and identifying **peer references**.

---

### **10. CCR Model â€“ Two Orientations**

There are two main forms of the CCR model:

| **Type**                      | **Focus**                                      | **Goal**                                 |
| ----------------------------- | ---------------------------------------------- | ---------------------------------------- |
| **Input-Oriented CCR Model**  | Minimize inputs while keeping outputs constant | Used when inputs can be controlled       |
| **Output-Oriented CCR Model** | Maximize outputs while keeping inputs constant | Used when output improvement is the goal |

**Example:**

* A hospital (Input-Oriented): Try to treat the same number of patients with fewer staff/resources.
* A factory (Output-Oriented): Try to produce more products with the same resources.

---

### **11. Steps in Applying the CCR Model**

1. **Identify DMUs** (similar units to compare)
2. **Collect data** on inputs and outputs
3. **Choose orientation** (input or output)
4. **Use DEA/CCR model** (linear programming)
5. **Get efficiency scores** for all DMUs
6. **Identify efficient and inefficient DMUs**
7. **Find peer groups and targets** for inefficient ones
8. **Analyze results and plan improvements**

---

### **12. Example Scenario**

Letâ€™s take 3 bank branches (DMUs):

| Branch | Inputs (Employees) | Output (Transactions) |
| ------ | ------------------ | --------------------- |
| A      | 10                 | 900                   |
| B      | 12                 | 1200                  |
| C      | 10                 | 1200                  |

Here, Branch **C** is most efficient because it produces **1200** outputs with **10** inputs.
Branch A produces less with the same inputs â†’ **inefficient**.
So, C becomes **peer** for A, and Aâ€™s **target** is to reach 1200 transactions.

---

### **13. Advantages of the CCR Model**

âœ… Handles **multiple inputs and outputs** simultaneously  
âœ… Provides **relative efficiency** (not absolute)  
âœ… Identifies **benchmarks and targets**  
âœ… Supports **objective decision-making**  
âœ… Useful in both **public and private sectors**  

---

### **14. Limitations**

âŒ Assumes **constant returns to scale (CRS)** â€” doubling input doubles output (not always realistic)  
âŒ Sensitive to **data errors or missing data**  
âŒ Does not tell **why** inefficiency occurs â€” only how much  
âŒ All DMUs must be **comparable and homogeneous**

# ðŸ“Š **Unit 4 â€“ Part 3: Cross Efficiency Analysis (Virtual Inputs and Outputs)**

---

### **1. Introduction**

After using the **CCR model**, we know which DMUs (Decision-Making Units) are efficient and which are not.
But sometimes, **many DMUs get the same score (1 or 100%)**, and we need a better way to **differentiate** among them.

Thatâ€™s where **Cross Efficiency Analysis** comes in.

Itâ€™s a **more advanced DEA technique** that evaluates the efficiency of each DMU **not only by its own chosen weights** but also by the **weights of other DMUs**.

This gives a **fairer, more discriminative, and comprehensive** efficiency evaluation.

---

### **2. What is Cross Efficiency Analysis?**

Cross Efficiency Analysis is an **extension of the CCR Model** that:

* Removes bias from self-evaluation.
* Uses **peer evaluation** (how other DMUs perform using the same weights).
* Produces **virtual inputs and outputs** for comparison.

So instead of each DMU saying, â€œIâ€™m efficient with my own weights,â€
itâ€™s evaluated again using the weights of all others to see if it remains efficient.

---

### **3. Why Itâ€™s Needed**

In the normal CCR model:

* Each DMU chooses weights that **maximize its own efficiency**.
* As a result, many DMUs can become **efficient (score = 1)** even though some are slightly better than others.

Cross Efficiency fixes this by:

* Evaluating each DMU from **multiple perspectives (self and peers)**.
* Providing a more **balanced and reliable efficiency score**.

---

### **4. How It Works â€“ Concept Overview**

1. In the CCR model, we calculate:
   $$
   \text{Efficiency}*k = \frac{\sum (u_r \times y*{rk})}{\sum (v_i \times x_{ik})}
   $$
   where

   * $u_r$ = weight of output *r*
   * $v_i$ = weight of input *i*
   * $y_{rk}$ = output value of DMU *k*
   * $x_{ik}$ = input value of DMU *k*

2. Each DMU chooses its own $u_r$ and $v_i$ to maximize efficiency.

3. In Cross Efficiency, we take those weights from each DMU and apply them to **all other DMUs**.

4. The **average efficiency** across all evaluations (self + peers) becomes the **cross-efficiency score**.

---

### **5. Example to Understand**

Letâ€™s say there are **3 DMUs** â€” A, B, and C.

| DMU | Efficiency (Self-Evaluation) | Efficiency under Aâ€™s Weights | Efficiency under Bâ€™s Weights | Efficiency under Câ€™s Weights | **Cross Efficiency (Average)** |
| --- | ---------------------------- | ---------------------------- | ---------------------------- | ---------------------------- | ------------------------------ |
| A   | 1.00                         | 1.00                         | 0.85                         | 0.90                         | (1 + 0.85 + 0.90)/3 = 0.92     |
| B   | 1.00                         | 0.88                         | 1.00                         | 0.92                         | (0.88 + 1 + 0.92)/3 = 0.93     |
| C   | 1.00                         | 0.86                         | 0.90                         | 1.00                         | (0.86 + 0.90 + 1)/3 = 0.92     |

Even though all three had 1.00 in self-evaluation,
Cross Efficiency shows that **DMU B (0.93)** performs best when judged by all others.

---

### **6. Virtual Inputs and Outputs**

In Cross Efficiency, when weights are applied to other DMUs, we generate **virtual (weighted) inputs and outputs**.

They are called **virtual** because they represent **weighted values**, not actual ones.

* **Virtual Input:** $v_1 x_1 + v_2 x_2 + ... + v_n x_n$
* **Virtual Output:** $u_1 y_1 + u_2 y_2 + ... + u_m y_m$

These help in comparing DMUs on a **common, standardized scale** â€” because actual inputs and outputs might have different units (money, hours, number of employees, etc.).

---

### **7. Steps in Cross Efficiency Analysis**

| **Step**                                | **Description**                                                         |
| --------------------------------------- | ----------------------------------------------------------------------- |
| **1. Compute CCR efficiency**           | For each DMU using DEA (self-evaluation).                               |
| **2. Determine optimal weights (u, v)** | Each DMU chooses weights that maximize its own efficiency.              |
| **3. Apply weights to all DMUs**        | Evaluate each DMU using the weights of every other DMU.                 |
| **4. Create Cross Efficiency Matrix**   | Contains efficiency of each DMU under all weight sets.                  |
| **5. Calculate Average Score**          | For each DMU, take average of all evaluations â†’ cross-efficiency score. |

---

### **8. Benefits of Cross Efficiency Analysis**

âœ… Removes **selfish bias** in efficiency measurement.
âœ… Provides **better ranking** among efficient DMUs.
âœ… Encourages **peer learning** (compare against othersâ€™ criteria).
âœ… Identifies **robustly efficient** DMUs (those that perform well under multiple weight sets).
âœ… Improves **discrimination power** of DEA.

---

### **9. Limitations**

âŒ Computationally more complex (especially for large datasets).
âŒ Choice of averaging method (arithmetic, geometric, etc.) can affect results.
âŒ Still assumes data accuracy â€” sensitive to outliers or noise.
âŒ May reduce fairness if peer weights are too different.

---

**Scenario:** Evaluating 10 hospital branches.

# âš™ï¸ **Unit 4 â€“ Part 4: Other Models (Efficiency Measures)**

After the **CCR model** (Charnes, Cooper, and Rhodes) and **Cross Efficiency Analysis**, there are several **other models** used to measure and improve efficiency in organizations.
These models are mainly **extensions or variations of DEA (Data Envelopment Analysis)** that deal with **real-world challenges** like variable returns, slacks, non-radial inputs/outputs, etc.

Letâ€™s go through them one by one.

---

## **1. BCC Model (Banker, Charnes, and Cooper Model)**

### **Definition:**

The **BCC model** is an extension of the **CCR model** that allows for **Variable Returns to Scale (VRS)**.

In the **CCR model**, itâ€™s assumed that outputs will increase in direct proportion to inputs (Constant Returns to Scale).
But in real life, this is rarely true â€” sometimes increasing inputs doesnâ€™t increase output equally.

So, the **BCC model** relaxes that assumption and allows the relationship to vary.

---

### **Formula and Concept:**

* CCR model assumes:
  $$ \text{Output increases proportionally with Input} $$
  (Constant Returns to Scale)

* BCC model assumes:
  $$ \text{Output increases non-proportionally with Input} $$
  (Variable Returns to Scale)

This means:

* Some Decision Making Units (DMUs) may operate under **increasing returns to scale (IRS)** (efficiency improves as size grows)
* Others under **decreasing returns to scale (DRS)** (efficiency reduces as size grows)

---

### **Objective:**

To find whether inefficiency is due to **managerial performance** or **scale of operations**.

### **Use Case Example:**

Suppose we compare 10 hospitals.
A small hospital might perform better per resource used, even if its total output is smaller â€” BCC helps capture that efficiency more accurately than CCR.

---

## **2. Additive Model**

### **Definition:**

The **Additive Model** is a **non-radial** model in DEA, meaning it looks at the **sum of input excesses and output shortfalls** rather than proportional changes.

### **Concept:**

Instead of reducing all inputs or increasing all outputs by the same ratio, this model focuses on **individual slacks** (unused inputs or missed outputs).

### **Formula:**

$$ \text{Maximize:} \quad \sum (\text{Output Slacks}) + \sum (\text{Input Slacks}) $$

It helps in identifying where exactly inefficiencies lie â€” which input is being wasted or which output can be improved.

---

### **Use Case:**

If a bank uses more employees (input) but processes fewer loans (output), this model shows **how much input can be reduced or how much output should increase** to reach efficiency.

---

## **3. Slack-Based Measure (SBM) Model**

### **Definition:**

The **Slack-Based Measure (SBM)** model directly measures efficiency by incorporating **input and output slacks** into the efficiency score.

It gives a **true efficiency value** between 0 and 1, considering all unused resources and missing outputs.

---

### **Formula (Simplified):**

$$ \text{Efficiency} = 1 - \frac{\text{Average of input slacks}}{\text{Total inputs}} - \frac{\text{Average of output slacks}}{\text{Total outputs}} $$

So, if slacks are large, efficiency decreases.

---

### **Benefits:**

* More accurate than CCR or BCC when data has slacks.
* Suitable when efficiency cannot be captured proportionally.
* Reflects real operational inefficiencies.

---

## **4. Super-Efficiency Model**

### **Definition:**

The **Super-Efficiency Model** is used to **rank efficient DMUs** that all score 1 in regular DEA models.

In CCR or BCC, multiple DMUs may appear â€œefficientâ€ (efficiency = 1).
To differentiate them, the Super-Efficiency model removes the DMU under evaluation from the reference set and measures how efficient it still is compared to others.

---

### **Use Case Example:**

If 5 factories all show efficiency = 1 under DEA, the Super-Efficiency model will show values like:

* Factory A = 1.20
* Factory B = 1.10
* Factory C = 0.98

This helps rank the best among efficient ones.

---

## **5. Window Analysis Model**

### **Definition:**

**Window Analysis** evaluates DMUs **over time** rather than at a single point.

It treats the same DMU (say, a company or department) in different time periods as **different entities** and compares their performance trends.

---

### **Example:**

Suppose you analyze the efficiency of 5 hospitals from 2020 to 2025.
The same hospital in 2020, 2021, etc., will be considered as different DMUs.
This shows **improvement or decline** in efficiency over time.

---

### **Advantages:**

* Tracks performance trends.
* Shows which DMU is improving or declining.
* Useful for **time-series efficiency evaluation**.

---

## **6. Two-Stage DEA Model**

### **Definition:**

In some cases, efficiency cannot be measured in a single step because output from one stage becomes **input for another stage**.
The **Two-Stage DEA Model** handles such multi-stage processes.

---

### **Example:**

In a **university**:

* **Stage 1:** Inputs (faculty, funds) produce intermediate outputs (research papers).
* **Stage 2:** Intermediate outputs become inputs for final outputs (student placements, patents).

So, Two-Stage DEA measures the efficiency of both stages separately and overall.

---

## **7. Network DEA Model**

### **Definition:**

A **Network DEA Model** analyzes systems where multiple interrelated processes exist, like a **supply chain**, **production network**, or **multi-department organization**.

It treats each internal process as a â€œnodeâ€ in a network and evaluates the overall system efficiency.

---

### **Example:**

In a manufacturing company:

* Stage 1: Procurement
* Stage 2: Production
* Stage 3: Distribution

Each stage has inputs and outputs, and their interaction determines the total efficiency.

---

## **8. Stochastic DEA Model**

### **Definition:**

This model considers **uncertainty or randomness** in input/output data (like market fluctuations or random errors).

Itâ€™s a mix of **DEA** and **Statistical Analysis**, making it more realistic for industries where data changes frequently.

---

### **Use Case:**

Used in finance, agriculture, or energy sectors where environmental or market conditions affect efficiency.

---

## **9. Bootstrapped DEA Model**

### **Definition:**

The **Bootstrapped DEA** model applies **resampling techniques (bootstrap methods)** to correct for bias in efficiency scores and improve reliability.

This helps determine **confidence intervals** and **statistical significance** of efficiency results.

## **2. Types of Pattern Matching Techniques**

There are mainly two techniques used:

1. **Cluster Analysis** â€“ groups similar units together.
2. **Outlier Analysis** â€“ identifies extreme or abnormal units.

---

## **3. Cluster Analysis**

### **Definition:**

**Cluster Analysis** is a statistical method used to **group objects (DMUs)** so that:

* Units in the **same cluster** are more similar to each other.
* Units in **different clusters** are more different from each other.

In short:

> Itâ€™s like organizing students by performance â€” high, average, and low achievers â€” based on similar marks or efficiency.

---

### **3.1 Objectives of Cluster Analysis**

* To identify **peer groups** of DMUs.
* To understand **efficiency patterns** among similar types of units.
* To help managers **benchmark** against similar organizations.
* To support **strategic decisions** by analyzing group behavior.

---

### **3.2 Steps in Cluster Analysis**

1. **Select Variables:** Choose performance indicators (inputs/outputs or efficiency scores).
2. **Measure Similarity:** Calculate how similar or different DMUs are (using distance measures).
3. **Form Clusters:** Use algorithms to form groups of similar units.
4. **Interpret Clusters:** Analyze what each cluster represents (efficient, moderate, or inefficient).
5. **Validation:** Check if the clusters make practical sense.

---

### **3.3 Common Methods of Clustering**

| **Method**                  | **Description**                                                              |
| --------------------------- | ---------------------------------------------------------------------------- |
| **Hierarchical Clustering** | Builds a tree-like structure (dendrogram) to show how units are grouped.     |
| **K-Means Clustering**      | Divides data into *k* clusters based on distance from the center (centroid). |
| **DBSCAN**                  | Groups based on density; separates noisy or scattered data points.           |
| **Fuzzy C-Means**           | Allows a DMU to belong to multiple clusters with probabilities.              |

---

### **3.4 Example**

Suppose we have 10 hospitals with efficiency scores:

| Hospital | Efficiency |
| -------- | ---------- |
| H1       | 0.95       |
| H2       | 0.92       |
| H3       | 0.90       |
| H4       | 0.70       |
| H5       | 0.68       |
| H6       | 0.50       |
| H7       | 0.48       |
| H8       | 0.96       |
| H9       | 0.47       |
| H10      | 0.69       |

After cluster analysis:

* **Cluster 1 (Efficient):** H1, H2, H3, H8
* **Cluster 2 (Moderate):** H4, H5, H10
* **Cluster 3 (Inefficient):** H6, H7, H9

Now, management can study why Cluster 1 hospitals perform better and guide others accordingly.

---

### **3.5 Applications of Cluster Analysis in Efficiency**

* Grouping banks or hospitals based on DEA efficiency results.
* Identifying **operational similarities** or **common practices**.
* Supporting **benchmarking** â€” finding the most relevant peers.
* Detecting **regional or structural performance patterns**.

---

## **4. Outlier Analysis**

### **Definition:**

**Outlier Analysis** identifies **unusual or exceptional DMUs** whose performance deviates significantly from others.

In efficiency measurement, an **outlier** could be:

* A **super-efficient DMU** (performing far better than average), or
* A **poorly performing DMU** (way below average).

---

### **4.1 Why Outlier Detection Is Important**

* To **avoid distortion** in efficiency scores.
* To **understand special cases** (e.g., a company with unique strategies).
* To **remove noise** or **erroneous data** that may mislead results.
* To **improve data quality** and **reliability** of models.

---

### **4.2 Causes of Outliers**

| **Cause**                       | **Example**                                              |
| ------------------------------- | -------------------------------------------------------- |
| **Data Entry Errors**           | Wrongly recorded input/output data                       |
| **Exceptional Efficiency**      | A hospital using AI-based diagnosis outperforming others |
| **Unique Operating Conditions** | A company in a different market or region                |
| **Measurement Issues**          | Using different scales or units of measurement           |

---

### **4.3 Methods of Outlier Detection**

| **Method**                   | **Description**                                                            |
| ---------------------------- | -------------------------------------------------------------------------- |
| **Visual Methods**           | Scatter plots, box plots, or histograms to spot extreme points.            |
| **Statistical Methods**      | Z-score, IQR (Interquartile Range), or standard deviation-based detection. |
| **DEA-based Detection**      | Checking DMUs with extremely high/low efficiency or slacks.                |
| **Distance-based Detection** | Using Euclidean or Mahalanobis distance from group centers.                |

---

### **4.4 Example**

In a DEA analysis of 15 schools:

* 12 schools have efficiency between **0.6 and 0.9**
* 2 schools have efficiency **above 1.2**
* 1 school has efficiency **below 0.3**

These 3 are **outliers** â€” their operations differ greatly.
Analysts will then verify if this difference is due to:

* Wrong data,
* Special technology use, or
* Exceptional management.

---

### **4.5 Handling Outliers**

* **Verify Data:** Check for input/output accuracy.
* **Exclude Temporarily:** Run DEA without outliers to check model stability.
* **Analyze Separately:** Study them as special cases.
* **Adjust Model:** Use robust DEA models (like Bootstrapped DEA).

---

## **5. Relationship Between Cluster and Outlier Analysis**

| **Aspect**  | **Cluster Analysis**       | **Outlier Analysis**           |
| ----------- | -------------------------- | ------------------------------ |
| **Purpose** | Group similar DMUs         | Identify unusual DMUs          |
| **Focus**   | Similarity patterns        | Deviations or exceptions       |
| **Result**  | Clusters (groups)          | Isolated or special points     |
| **Use**     | Benchmarking, segmentation | Data validation, anomaly study |

Both are complementary â€”
**Cluster Analysis** finds **patterns**,
**Outlier Analysis** finds **exceptions**.
