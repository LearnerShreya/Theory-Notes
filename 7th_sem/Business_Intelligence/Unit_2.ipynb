# **Unit 2: Knowledge Delivery**

### **Definition of Knowledge Delivery**

**Knowledge Delivery** in BI refers to the process of **delivering the right information, to the right users, at the right time, in the right format** to support decision-making.

It is the **final stage of the BI process**, where insights gained from data analysis are presented to different types of users (e.g., executives, analysts, operational staff, external users) through reports, dashboards, alerts, queries, and other tools.

üëâ In simple words:
**Knowledge Delivery = How BI results (reports, dashboards, alerts) are delivered to users so they can make effective decisions.**

## **1. Business Intelligence User Types**

Different BI users use data in different ways:

### (a) **Operational Users**

* **Who**: Frontline staff, customer service reps, sales associates.
* **Needs**: Quick and simple dashboards, real-time reports.
* **Tools**: Predefined reports, mobile BI, embedded BI.
* **Goal**: Support **day-to-day operations** (e.g., sales tracking, customer handling).
* **Example**: A cashier checking real-time stock before billing.

---

### (b) **Power Users / Analysts**

* **Who**: Business analysts, data scientists, advanced users.
* **Needs**: Ability to explore data, run custom queries, predictive models.
* **Tools**: SQL, Tableau, Power BI, data mining software.
* **Goal**: Perform **deep analysis** and find insights.
* **Example**: Analyst identifying why sales dropped in a specific region.

---

### (c) **Executives & Managers**

* **Who**: CEOs, department heads, managers.
* **Needs**: High-level KPIs, trend analysis, summary dashboards.
* **Tools**: Executive dashboards, scorecards, alerts.
* **Goal**: Support **strategic and tactical decision-making**.
* **Example**: CEO checking overall company revenue trend before planning expansion.

---

### (d) **IT & Data Specialists**

* **Who**: Database admins, IT staff, data engineers.
* **Needs**: Manage pipelines, ensure security, maintain BI systems.
* **Tools**: Data warehouses, ETL tools, cloud BI platforms.
* **Goal**: Ensure **data integrity, governance, and system performance**.
* **Example**: IT team setting up secure data storage for BI tools.

---

### (e) **External Users**

* **Who**: Customers, partners, suppliers.
* **Needs**: Limited but relevant data (e.g., order tracking, supply chain performance).
* **Tools**: Customer portals, supplier dashboards.
* **Goal**: Improve **collaboration and transparency**.
* **Example**: Supplier checking demand forecasts via BI dashboard.

## **2. Standard Reports**

* **Definition**: Predefined, recurring reports presenting structured information.
* **Examples**: Monthly sales report, quarterly financial summary, HR headcount report.

### Characteristics:

* Fixed format and predefined structure.
* Generated regularly (daily, weekly, monthly).
* Focus on **KPIs**.
* Limited customization by end users.
* Easy to share and understand.

### Importance:

* Ensures **consistency** across departments.
* Useful for **regulatory and compliance reporting**.
* Provides quick business snapshots.
* Helps in **trend analysis and historical comparisons**.

### Benefits:

* Reliable (everyone sees the same data).
* Automated (saves manual work).
* Comparable (supports benchmarking).
* Accessible to non-technical users.

### Limitations:

* Lack of flexibility for deeper exploration.
* Might not answer ad-hoc business questions.
* Risk of information overload.
* Dependent on initial report design.

## **3. Interactive Analysis & Ad-Hoc Querying**

### **Interactive Analysis**

* **Definition**: Exploring data dynamically using drill-down, filters, and pivoting.
* **Features**:

  * Drill-down/roll-up.
  * Filtering, slicing/dicing.
  * Visualization through charts & dashboards.
  * ‚ÄúWhat-if‚Äù scenario modeling.
* **Benefits**:

  * Faster decisions.
  * Flexible exploration.
  * Encourages insights and data discovery.
  * Reduces IT dependency.
* **Example**: Manager drilling down from total sales ‚Üí by region ‚Üí by product.

---

### **Ad-Hoc Querying**

* **Definition**: On-demand custom queries created by users for specific questions.
* **Features**:

  * Drag-and-drop interfaces.
  * SQL-like queries for advanced users.
  * Real-time database access.
  * Ability to save/share queries.
* **Benefits**:

  * Flexibility (tailored reports).
  * Speed (immediate answers).
  * Empowers business users.
  * Supports **unique/non-routine questions**.
* **Example**: Marketing analyst querying ‚ÄúTop 10 customers in Q2 by revenue and location.‚Äù

## **4. Parameterized Reports & Self-Service Reporting**

### **Parameterized Reports**

* **Definition**: Reports that allow user input (parameters) to customize output.
* **Example**: Sales report by selecting **region** or **time period**.
* **Benefits**:

  * Flexibility ‚Äì tailor to user‚Äôs needs.
  * Reusability ‚Äì one report, multiple uses.
  * Improved analysis ‚Äì view data from multiple angles.
  * Automation ‚Äì works in scheduled systems.
  * Security ‚Äì limit access via user-specific parameters.
* **Tools**: Power BI, Tableau, SSRS, Crystal Reports.

---

### **Self-Service Reporting**

* **Definition**: BI approach that lets users **create their own reports** without IT help.
* **Goal**: Make BI more **accessible, faster, and flexible**.
* **Benefits**:

  1. **Productivity** ‚Äì quicker report creation.
  2. **Faster decisions** ‚Äì no IT delays.
  3. **Higher adoption** ‚Äì more users engage with BI.
* **Tools**: Tableau, Power BI, QlikView, Looker, SAP Analytics Cloud, Zoho Analytics.
* **Example**: Department manager creating a performance dashboard without IT support.

---

## **5. Dimensional Analysis**

* **Definition**: Analyzing data from different perspectives (dimensions).
* **Key Concepts**:

  * **Facts**: Quantitative data (sales, revenue, profit).
  * **Dimensions**: Descriptive context (time, product, customer, location).
  * **Fact Table**: Stores numeric measures (e.g., sales amount).
  * **Dimension Tables**: Describe ‚Äúwho, what, where, when.‚Äù
* **Benefits**:

  * Improved insights (multi-perspective view).
  * Faster querying.
  * Better decisions through comprehensive view.
* **Examples**:

  * Sales by **region, product, and time**.
  * Customer analysis by **demographics and preferences**.
* **Tools**: OLAP tools (Oracle OLAP, MS Analysis Services, IBM Cognos), Data Warehouses (Redshift, BigQuery, Snowflake).

---

## **6. Alerts / Notifications**

* **Definition**: Automated BI messages triggered by events, thresholds, or anomalies.
* **Types**:

  * **Threshold alerts** ‚Äì e.g., sales drop below target.
  * **Trend alerts** ‚Äì e.g., continuous revenue decline.
  * **Anomaly alerts** ‚Äì e.g., sudden unusual transactions.
* **Purpose**:

  * Real-time awareness.
  * Proactive decision-making.
  * Performance monitoring.
  * Exception reporting (fraud detection, errors).
* **Benefits**:

  * Proactive actions.
  * Improved monitoring.
  * Increased efficiency.
* **Tools**: Tableau, Power BI, QlikView, Splunk, Datadog, Nagios, Prometheus.
