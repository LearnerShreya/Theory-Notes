## **Unit 1: Business Intelligence**

* **Effective and Timely Decisions** – Importance of making decisions based on accurate and up-to-date data.
* **Data, Information, and Knowledge** – Transformation from raw data → meaningful information → actionable knowledge.
* **Role of Mathematical Models** – Use of models and algorithms for prediction, optimization, and analysis.
* **Business Intelligence Architectures** – Frameworks and systems that support BI processes.
* **Cycle of a Business Intelligence Analysis** – Data collection → processing → analysis → insights → decisions.
* **Enabling Factors in Business Intelligence Projects** – Technology, skilled people, organizational support, quality data.
* **Development of a Business Intelligence System** – Steps to design, build, and deploy BI solutions.
* **Ethics and Business Intelligence** – Responsible data usage, privacy, fairness, and transparency.

---

## **Unit 2: Knowledge Delivery**

* **Business Intelligence User Types** – Categories of users (operational, power users, executives, analysts).
* **Standard Reports** – Predefined, routine reports for regular decision-making.
* **Interactive Analysis and Ad Hoc Querying** – On-demand queries and interactive data exploration.
* **Parameterized Reports and Self-Service Reporting** – Reports customized by user inputs; self-service tools for non-technical users.
* **Dimensional Analysis** – Multidimensional view of data (OLAP – slice, dice, drill-down, roll-up).
* **Alerts / Notifications** – Automated messages and warnings based on predefined conditions or thresholds.
