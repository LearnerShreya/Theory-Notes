# ðŸ”¹ Part 6: Top-Down Parsing Techniques

Top-down parsing = build **parse tree from the root (start symbol) down to the leaves** (tokens).
ðŸ‘‰ Parser starts with the start symbol and keeps expanding using grammar rules until input tokens are matched.

---

## ðŸ“Œ 6.1 General Strategy

* Start with **Start Symbol (S)**.
* Repeatedly replace non-terminals using production rules.
* Try to **match input tokens** from **left to right**.
* If successful â†’ string accepted.
* If mismatch â†’ error.

Two main approaches:

1. **Recursive Descent Parsing** (backtracking or non-backtracking)
2. **LL(1) Parsing** (Predictive parser without backtracking)

---

## ðŸ“Œ 6.2 Recursive Descent Parser

* **Manual implementation** of grammar using recursive functions.
* Each **non-terminal â†’ one function**.
* Each **production â†’ coded as part of that function**.

### Example: Grammar

```
E â†’ T E'
E' â†’ + T E' | Îµ
T â†’ F T'
T' â†’ * F T' | Îµ
F â†’ (E) | id
```

### Parser Functions (pseudo-code):

```c
void E() {
    T();
    Eprime();
}

void Eprime() {
    if (lookahead == '+') {
        match('+');
        T();
        Eprime();
    }
    else return; // Îµ-production
}

void T() {
    F();
    Tprime();
}

void Tprime() {
    if (lookahead == '*') {
        match('*');
        F();
        Tprime();
    }
    else return; // Îµ-production
}

void F() {
    if (lookahead == '(') {
        match('(');
        E();
        match(')');
    }
    else if (lookahead == 'id') match('id');
    else error();
}
```

## 1. ðŸ“Œ Why SLR?

# ðŸ”¹ Part 7.6: **LALR Parsing**

---

## 1. ðŸ“Œ Why LALR?

* **Canonical LR (CLR)** is very powerful but creates **too many states** (huge parsing tables).
* **LALR (LookAhead LR)** reduces the table size while keeping almost the same power.
* Thatâ€™s why **most modern compilers (like YACC)** use **LALR parsers**.

---

## 2. ðŸ“Œ Idea Behind LALR

* Start with the **Canonical LR(1) automaton**.
* **Merge states** that have the same **LR(0) items** (ignore lookahead).
* Keep **combined lookahead sets**.
* Result: fewer states, much smaller tables.

ðŸ‘‰ Example:

* CLR may produce 1000 states.
* LALR reduces to \~100 states.
* Yet parsing power is almost same (except a few rare ambiguous grammars).

---

## 3. ðŸ“Œ Steps to Construct LALR Parsing Table

1. **Construct LR(1) Item Sets** (as in CLR).
2. **Find states with same LR(0) core** (same productions, dot positions).
3. **Merge them** into a single state, combining their lookaheads.
4. **Build ACTION and GOTO tables** using merged states.

---

## 4. ðŸ“Œ Example (Simplified)

Grammar:

```
S' â†’ S
S  â†’ L = R | R
L  â†’ * R | id
R  â†’ L
```

### Canonical LR (CLR)

* Generates \~12 states.

### LALR

* Some states have same LR(0) core (only differ by lookahead).
* After merging â†’ \~7 states.
* Table becomes much smaller.

---

| State | ACTION (id, \*, =, \$) | GOTO (S, L, R)   |
| ----- | ---------------------- | ---------------- |
| I0    | id â†’ shift, \* â†’ shift | Sâ†’I1, Lâ†’I2, Râ†’I3 |
| I1    | \$ â†’ accept            |                  |
| I2    | = â†’ shift              | Râ†’I4             |

| â€¦     | â€¦                      | â€¦                |

## 6. ðŸ“Œ Comparison of LR Parsers

| Feature          | LR(0) | SLR         | LALR              | CLR (LR(1))        |
| ---------------- | ----- | ----------- | ----------------- | ------------------ |
| Lookahead        | None  | FOLLOW sets | Merged lookaheads | Precise (per item) |
| Table Size       | Small | Small       | Medium            | Very Large         |

# ðŸ”¹ Part 8: Error Handling and Recovery in Syntax Analyzer

We already discussed error handling in **general parsing** earlier, but here weâ€™ll look at **parser-specific error recovery** (important in exams).

---

## ðŸ“Œ Role of Error Recovery in Parsing

* Parsers must not stop at the **first syntax error**.
* They should **recover gracefully** so the compiler can find more errors in one run.
* Goal: **Provide meaningful error messages** and **continue parsing**.

---

## ðŸ“Œ Techniques of Error Recovery in Parsers

### 1. **Panic Mode Recovery**

* On error, parser **discards input symbols** until a synchronizing token is found (`;`, `}`, etc.).
* Example:

  ```
  if (x < y { z = 10; }
  ```

  Missing `)` â†’ parser skips tokens until `;` or `}`.
* âœ… Simple, widely used.
* âŒ May skip too much code.

---

### 2. **Phrase-Level Recovery**

* Parser makes a **local correction** like inserting/deleting tokens.
* Example:
  Input: `if (x < y { z = 10; }`
  Parser **inserts `)`** automatically.
* âœ… Easy to implement.
* âŒ Might misinterpret the programmerâ€™s intent.

---

### 3. **Error Productions**

* Grammar is extended to include **common mistakes**.
* Example:

  ```
  stmt â†’ if (expr) stmt
       | if (expr) stmt else stmt
       | if expr stmt   â† error production
  ```
* âœ… Allows customized error messages.

---

### 4. **Global Correction**

* Compiler tries to compute the **minimum changes** required to make program valid.
* Example: Insert/delete a few tokens to continue parsing.
* âœ… Gives best results.
* âŒ Very expensive (not practical).

---

## ðŸ“Œ Error Handling in Different Parsers

* **LL(1) Parser** â†’ Panic mode (skip until synchronizing set).
* **LR Parser** â†’ Error entries in parsing table + phrase-level corrections.

---

âœ… **Summary**:

* Error handling ensures smooth compilation.
* Four main strategies: **Panic mode, Phrase-level, Error productions, Global correction**.
* **Panic mode** is most practical & widely used.

# ðŸ”¹ Part 9: YACC Tool (Yet Another Compiler Compiler)

## ðŸ“Œ What is YACC?

* **YACC** = "Yet Another Compiler Compiler".
* It is a **parser generator tool**.

---

## ðŸ“Œ Role of YACC

* Generates a **parser in C language** automatically from grammar rules.
* Helps in **syntax analysis and error detection**.
* Works on **LALR(1) parsing technique**.

---

## ðŸ“Œ Structure of YACC Program

A YACC program has **3 sections** separated by `%%`:

```
Declarations
%%
Rules
%%
Auxiliary functions
```

### 1. **Declarations Section**

* Define tokens (from Lex).
* Define variables and data types.

Example:

```yacc
%token NUM ID
%left '+' '-'
%left '*' '/'
```

---

### 2. **Rules Section**

* Contains **grammar rules** + associated **actions (C code)**.
* Format:

  ```
  LHS : RHS { action }
  ```
* Example:

  ```yacc
  expr : expr '+' expr { printf("Add\n"); }
       | expr '*' expr { printf("Multiply\n"); }
       | '(' expr ')'  { }
       | NUM           { printf("Number\n"); }

  ```

---

### 3. **Auxiliary Functions**

* User-defined helper functions (like `yyerror()`, `main()`).

Example:

```c
int main() {
   yyparse();  // Calls parser
   return 0;
}

void yyerror(char *s) {
   printf("Error: %s\n", s);
}
```

## ðŸ“Œ Working of Lex + YACC

1. **Lex file** generates `lex.yy.c` â†’ performs lexical analysis.
2. **YACC file** generates `y.tab.c` â†’ performs syntax analysis.
